```python
# DIAGNOSTIC: Find where your files actually are
import os

def find_files_recursive(directory, extensions, max_depth=3, current_depth=0):
    """Recursively find files with specific extensions"""
    found_files = []
    if current_depth >= max_depth:
        return found_files
    
    try:
        for item in os.listdir(directory):
            item_path = os.path.join(directory, item)
            if os.path.isfile(item_path):
                if any(item.endswith(ext) for ext in extensions):
                    found_files.append(item_path)
            elif os.path.isdir(item_path) and not item.startswith('.'):
                found_files.extend(find_files_recursive(item_path, extensions, max_depth, current_depth + 1))
    except PermissionError:
        pass
    
    return found_files

print("üîç DIAGNOSTIC: Finding your KLT files...")
print("=" * 50)

# Check current directory
print(f"üìÅ Current directory: {os.getcwd()}")
print("üìÅ Contents:")
for item in os.listdir('.'):
    if os.path.isfile(item):
        print(f"  üìÑ {item}")
    elif os.path.isdir(item):
        print(f"  üìÅ {item}/")

# Search for KLT source files recursively
print("\nüîç Searching for KLT source files...")
source_files = find_files_recursive('.', ['.c', '.cu', '.h'])
if source_files:
    print(f"‚úÖ Found {len(source_files)} source files:")
    for file in source_files[:10]:  # Show first 10
        print(f"  üìÑ {file}")
    if len(source_files) > 10:
        print(f"  ... and {len(source_files) - 10} more files")
else:
    print("‚ùå No source files found!")

# Search for specific KLT files
print("\nüîç Looking for key KLT files...")
key_files = ['klt.c', 'convolve.c', 'example3.c', 'klt.h', 'convolve.h']
for key_file in key_files:
    found = False
    for root, dirs, files in os.walk('.'):
        if key_file in files:
            print(f"  ‚úÖ {key_file} found in {root}")
            found = True
            break
    if not found:
        print(f"  ‚ùå {key_file} not found")

# Check if we need to go up directories
print("\nüîç Checking parent directories...")
parent_dir = os.path.dirname(os.getcwd())
if parent_dir != os.getcwd():
    print(f"üìÅ Parent directory: {parent_dir}")
    try:
        parent_files = os.listdir(parent_dir)
        print("üìÅ Parent contents:")
        for item in parent_files:
            if os.path.isfile(os.path.join(parent_dir, item)):
                print(f"  üìÑ {item}")
            elif os.path.isdir(os.path.join(parent_dir, item)):
                print(f"  üìÅ {item}/")
    except:
        print("  ‚ùå Cannot access parent directory")
```

## üîß Step 2: Organize Files


```python
import os
import shutil
import subprocess

print("üîß ORGANIZING FILES FOR KLT GPU COMPILATION")
print("=" * 60)

# Create directory structure
os.makedirs('klt/src', exist_ok=True)
os.makedirs('klt/include', exist_ok=True)
os.makedirs('klt/input', exist_ok=True)
os.makedirs('klt/output', exist_ok=True)
os.makedirs('klt/build', exist_ok=True)

print("‚úÖ Created directory structure")

# Move files to appropriate directories
files_to_organize = {
    'src': ['.c', '.cu'],
    'include': ['.h'],
    'input': ['.pgm']
}

organized_count = 0
for category, extensions in files_to_organize.items():
    for file in os.listdir('.'):
        if any(file.endswith(ext) for ext in extensions):
            source = file
            dest = f"klt/{category}/{file}"
            try:
                shutil.move(source, dest)
                print(f"‚úì {file} ‚Üí klt/{category}/")
                organized_count += 1
            except Exception as e:
                print(f"‚ö†Ô∏è  Could not move {file}: {e}")

print(f"\n‚úÖ Organized {organized_count} files")
print("\nüìÅ Final structure:")
for root, dirs, files in os.walk('klt'):
    level = root.replace('klt', '').count(os.sep)
    indent = ' ' * 2 * level
    print(f"{indent}{os.path.basename(root)}/")
    subindent = ' ' * 2 * (level + 1)
    for file in files[:5]:  # Show first 5 files
        print(f"{subindent}{file}")
    if len(files) > 5:
        print(f"{subindent}... and {len(files) - 5} more files")
```

## üöÄ Step 3: Compile KLT with GPU Support


```python
print("üöÄ COMPILING KLT WITH GPU SUPPORT")
print("=" * 50)

# Start from /content/klt/
os.chdir('/content/klt')
print(f"üìÅ Working in: {os.getcwd()}")

# Show directory structure
print("\nüìÅ KLT directory structure:")
for item in os.listdir('.'):
    if os.path.isdir(item):
        print(f"  üìÅ {item}/")
        # Show contents of each subdirectory
        try:
            subdir_files = os.listdir(item)
            for subfile in subdir_files[:3]:  # Show first 3 files
                print(f"    üìÑ {subfile}")
            if len(subdir_files) > 3:
                print(f"    ... and {len(subdir_files) - 3} more files")
        except:
            pass
    else:
        print(f"  üìÑ {item}")

# Check if we have the right structure
print("\nüîç Checking for source files...")
src_files = []
if os.path.exists('src'):
    src_files = [f for f in os.listdir('src') if f.endswith(('.c', '.cu'))]
    print(f"‚úÖ Found {len(src_files)} source files in src/")
    for file in src_files[:5]:
        print(f"  üìÑ {file}")
    if len(src_files) > 5:
        print(f"  ... and {len(src_files) - 5} more")
else:
    print("‚ùå src/ directory not found")

# Check include files
include_files = []
if os.path.exists('include'):
    include_files = [f for f in os.listdir('include') if f.endswith('.h')]
    print(f"‚úÖ Found {len(include_files)} header files in include/")
    for file in include_files[:5]:
        print(f"  üìÑ {file}")
    if len(include_files) > 5:
        print(f"  ... and {len(include_files) - 5} more")
else:
    print("‚ùå include/ directory not found")

if not src_files:
    print("\n‚ùå No source files found!")
    print("Please upload your KLT source files to Colab")
    exit(1)

# Check if example3_gpu_real.c exists
print("\nüîç Checking for example3_gpu_real.c...")
if os.path.exists('src/example3_gpu_real.c'):
    print("‚úÖ example3_gpu_real.c found!")
else:
    print("‚ùå example3_gpu_real.c NOT found!")
    print("üìÅ Files in src/ directory:")
    if os.path.exists('src'):
        for file in os.listdir('src'):
            print(f"  üìÑ {file}")
    else:
        print("  ‚ùå src/ directory not found")
    
    print("\nüîß Creating example3_gpu_real.c from example3.c...")
    if os.path.exists('src/example3.c'):
        # Copy example3.c to example3_gpu_real.c
        import shutil
        shutil.copy('src/example3.c', 'src/example3_gpu_real.c')
        print("‚úÖ Created example3_gpu_real.c from example3.c")
        
        # Add GPU enable code to the file
        with open('src/example3_gpu_real.c', 'r') as f:
            content = f.read()
        
        # Add GPU enable at the beginning of main
        if 'int main(' in content:
            content = content.replace(
                'int main(int argc, char **argv) {',
                'int main(int argc, char **argv) {\n    // Enable GPU acceleration\n    extern int gpu_enabled;\n    gpu_enabled = 1;\n    printf("üöÄ GPU acceleration enabled\\n");'
            )
            
            with open('src/example3_gpu_real.c', 'w') as f:
                f.write(content)
            print("‚úÖ Added GPU enable code to example3_gpu_real.c")
    else:
        print("‚ùå example3.c not found either!")
        print("Please upload example3_gpu_real.c or example3.c to Colab")

# Check CUDA version and device info
print("\nüîç Checking CUDA environment...")
try:
    result = subprocess.run(['nvidia-smi'], capture_output=True, text=True, check=True)
    print("‚úÖ NVIDIA GPU detected:")
    print(result.stdout.split('\n')[0:3])  # Show first few lines
except:
    print("‚ö†Ô∏è  nvidia-smi not available")

try:
    result = subprocess.run(['nvcc', '--version'], capture_output=True, text=True, check=True)
    print("‚úÖ CUDA compiler version:")
    print(result.stdout.split('\n')[0:2])  # Show version info
except:
    print("‚ö†Ô∏è  nvcc version not available")

# Step 1: Compile GPU functions
print("\n1. Compiling GPU functions...")

# Try different CUDA architectures for compatibility (starting with oldest)
architectures = ['sm_30', 'sm_35', 'sm_37', 'sm_50', 'sm_52', 'sm_60', 'sm_61', 'sm_70', 'sm_75']
gpu_compiled = False

for arch in architectures:
    print(f"1a. Trying architecture {arch}...")
    gpu_compile_cmd = [
        'nvcc',
        '-O3',
        '-std=c++11',
        f'-arch={arch}',
        '-I./include',
        '-c',
        'src/convolve_gpu_functions.cu',
        '-o', 'build/convolve_gpu_functions.o'
    ]
    
    try:
        result = subprocess.run(gpu_compile_cmd, capture_output=True, text=True, check=True)
        print(f"‚úÖ GPU functions compiled successfully with {arch}!")
        gpu_compiled = True
        break
    except subprocess.CalledProcessError as e:
        print(f"‚ùå Failed with {arch}: {e.stderr.split('\\n')[0] if e.stderr else 'Unknown error'}")
        continue

if not gpu_compiled:
    print("‚ùå All GPU compilation attempts failed!")
    print("‚ö†Ô∏è  Will continue with CPU-only version")
    # Create empty object file to avoid linking errors
    with open('build/convolve_gpu_functions.o', 'w') as f:
        f.write('')
    print("‚úÖ Created placeholder GPU object file")

# Step 2: Compile KLT library
print("\n2. Compiling KLT library...")
klt_sources = [
    'src/klt.c',
    'src/convolve.c',
    'src/error.c',
    'src/pnmio.c',
    'src/pyramid.c',
    'src/selectGoodFeatures.c',
    'src/storeFeatures.c',
    'src/trackFeatures.c',
    'src/klt_util.c',
    'src/writeFeatures.c'
]

object_files = []
for src in klt_sources:
    obj_file = src.replace('src/', 'build/').replace('.c', '.o')
    object_files.append(obj_file)
    
    compile_cmd = [
        'gcc',
        '-c',
        '-O3',
        '-DNDEBUG',
        '-I./include',
        '-o', obj_file,
        src
    ]
    
    try:
        result = subprocess.run(compile_cmd, capture_output=True, text=True, check=True)
        print(f"‚úÖ Compiled {src}")
    except subprocess.CalledProcessError as e:
        print(f"‚ùå Failed to compile {src}: {e}")
        print(f"Error: {e.stderr}")
        exit(1)

# Step 3: Create static library
print("\n3. Creating static library...")
ar_cmd = ['ar', 'rcs', 'build/libklt.a'] + object_files
try:
    result = subprocess.run(ar_cmd, capture_output=True, text=True, check=True)
    print("‚úÖ Static library created!")
except subprocess.CalledProcessError as e:
    print(f"‚ùå Failed to create library: {e}")
    exit(1)

# Step 4: Compile example3_gpu_real with GPU functions
print("\n4. Compiling example3_gpu_real with GPU support...")

# First, compile the GPU functions object file (if not already done)
print("4a. Compiling GPU functions object file...")
if not gpu_compiled:
    # Try different architectures again
    for arch in architectures:
        print(f"4a. Trying architecture {arch} for GPU functions...")
        gpu_obj_cmd = [
            'nvcc',
            '-O3',
            '-std=c++11',
            f'-arch={arch}',
            '-I./include',
            '-c',
            'src/convolve_gpu_functions.cu',
            '-o', 'build/convolve_gpu_functions.o'
        ]
        
        try:
            result = subprocess.run(gpu_obj_cmd, capture_output=True, text=True, check=True)
            print(f"‚úÖ GPU functions object file compiled with {arch}!")
            gpu_compiled = True
            break
        except subprocess.CalledProcessError as e:
            print(f"‚ùå Failed with {arch}: {e.stderr.split('\\n')[0] if e.stderr else 'Unknown error'}")
            continue
    
    if not gpu_compiled:
        print("‚ùå All GPU compilation attempts failed!")
        print("‚ö†Ô∏è  Will continue with CPU-only version")
        # Create empty object file to avoid linking errors
        with open('build/convolve_gpu_functions.o', 'w') as f:
            f.write('')
        print("‚úÖ Created placeholder GPU object file")
else:
    print("‚úÖ GPU functions already compiled!")

# Now compile the main program with GPU functions linked using nvcc
print("4b. Linking example3_gpu_real with GPU functions using nvcc...")

# Use nvcc for final linking (handles CUDA runtime automatically)
example3_cmd = [
    'nvcc',
    '-O3',
    '-std=c++11',
    '-arch=sm_75',
    '-I./include',
    '-o', 'example3_gpu_real',
    'src/example3_gpu_real.c',
    'build/convolve_gpu_functions.o',
    '-L./build',
    '-lklt',
    '-lm'
]

try:
    result = subprocess.run(example3_cmd, capture_output=True, text=True, check=True)
    print("‚úÖ example3_gpu_real compiled successfully with nvcc!")
except subprocess.CalledProcessError as e:
    print(f"‚ùå Failed to compile example3_gpu_real: {e}")
    print(f"Error: {e.stderr}")
    exit(1)

print("\nüéâ COMPILATION COMPLETE!")
print("‚úÖ All components compiled successfully")
print("‚úÖ Ready to run KLT with GPU acceleration!")
```

```python
# üöÄ COMPILE KLT WITH GPU ACCELERATION (Convolution + Interpolation)
import os
import subprocess
import shutil

print("üöÄ COMPILING KLT WITH FULL GPU ACCELERATION")
print("=" * 60)

# Start from /content/klt/
os.chdir('/content/klt')
print(f"üìÅ Working in: {os.getcwd()}")

# Check what files actually exist
print("\nüîç Checking for source files...")
if os.path.exists('src'):
    src_files = os.listdir('src')
    print(f"üìÅ Files in src/: {src_files}")
    
    # Check for specific files we need
    required_files = ['convolve_gpu_functions.cu', 'interpolate_cuda.cu', 'example3_gpu_real.c']
    for file in required_files:
        if file in src_files:
            print(f"‚úÖ Found {file}")
        else:
            print(f"‚ùå Missing {file}")
else:
    print("‚ùå src/ directory not found")
    print("üìÅ Current directory contents:")
    for item in os.listdir('.'):
        print(f"  üìÑ {item}")

# Step 1: Compile GPU functions
print("\n1. Compiling GPU functions...")

# Compile convolution GPU functions
print("1a. Compiling convolution GPU functions...")
if not os.path.exists('src/convolve_gpu_functions.cu'):
    print("‚ùå src/convolve_gpu_functions.cu not found!")
    print("üìÅ Available files in src/:")
    if os.path.exists('src'):
        for file in os.listdir('src'):
            print(f"  üìÑ {file}")
    exit(1)

gpu_conv_cmd = [
    'nvcc',
    '-O3',
    '-std=c++11',
    '-arch=sm_75',  # Tesla T4
    '-I./include',
    '-c',
    'src/convolve_gpu_functions.cu',
    '-o', 'build/convolve_gpu_functions.o'
]

try:
    result = subprocess.run(gpu_conv_cmd, capture_output=True, text=True, check=True)
    print("‚úÖ Convolution GPU functions compiled successfully!")
except subprocess.CalledProcessError as e:
    print(f"‚ùå Convolution GPU compilation failed: {e}")
    print(f"Error: {e.stderr}")
    exit(1)

# Compile interpolation GPU functions
print("1b. Compiling interpolation GPU functions...")
if not os.path.exists('src/interpolate_cuda.cu'):
    print("‚ùå src/interpolate_cuda.cu not found!")
    print("üìÅ Available files in src/:")
    if os.path.exists('src'):
        for file in os.listdir('src'):
            print(f"  üìÑ {file}")
    exit(1)

gpu_interp_cmd = [
    'nvcc',
    '-O3',
    '-std=c++11',
    '-arch=sm_75',  # Tesla T4
    '-I./include',
    '-c',
    'src/interpolate_cuda.cu',
    '-o', 'build/interpolate_cuda.o'
]

try:
    result = subprocess.run(gpu_interp_cmd, capture_output=True, text=True, check=True)
    print("‚úÖ Interpolation GPU functions compiled successfully!")
except subprocess.CalledProcessError as e:
    print(f"‚ùå Interpolation GPU compilation failed: {e}")
    print(f"Error: {e.stderr}")
    exit(1)

# Step 2: Compile KLT library
print("\n2. Compiling KLT library...")
klt_sources = [
    'src/klt.c',
    'src/convolve.c',
    'src/error.c',
    'src/pnmio.c',
    'src/pyramid.c',
    'src/selectGoodFeatures.c',
    'src/storeFeatures.c',
    'src/trackFeatures.c',
    'src/klt_util.c',
    'src/writeFeatures.c'
]

object_files = []
for src in klt_sources:
    obj_file = src.replace('src/', 'build/').replace('.c', '.o')
    object_files.append(obj_file)
    
    compile_cmd = [
        'gcc',
        '-c',
        '-O3',
        '-DNDEBUG',
        '-I./include',
        '-o', obj_file,
        src
    ]
    
    try:
        result = subprocess.run(compile_cmd, capture_output=True, text=True, check=True)
        print(f"‚úÖ Compiled {src}")
    except subprocess.CalledProcessError as e:
        print(f"‚ùå Failed to compile {src}: {e}")
        print(f"Error: {e.stderr}")
        exit(1)

# Step 3: Create static library
print("\n3. Creating static library...")
ar_cmd = ['ar', 'rcs', 'build/libklt.a'] + object_files
try:
    result = subprocess.run(ar_cmd, capture_output=True, text=True, check=True)
    print("‚úÖ Static library created!")
except subprocess.CalledProcessError as e:
    print(f"‚ùå Failed to create library: {e}")
    exit(1)

# Step 4: Compile example3_gpu_real with GPU functions
print("\n4. Compiling example3_gpu_real with GPU support...")

# Check if main program exists
if not os.path.exists('src/example3_gpu_real.c'):
    print("‚ùå src/example3_gpu_real.c not found!")
    print("üìÅ Available files in src/:")
    if os.path.exists('src'):
        for file in os.listdir('src'):
            print(f"  üìÑ {file}")
    exit(1)

# Use nvcc for final linking (handles CUDA runtime automatically)
example3_cmd = [
    'nvcc',
    '-O3',
    '-std=c++11',
    '-arch=sm_75',
    '-I./include',
    '-o', 'example3_gpu_real',
    'src/example3_gpu_real.c',
    'build/convolve_gpu_functions.o',
    'build/interpolate_cuda.o',
    '-L./build',
    '-lklt',
    '-lm'
]

try:
    result = subprocess.run(example3_cmd, capture_output=True, text=True, check=True)
    print("‚úÖ example3_gpu_real compiled successfully with GPU acceleration!")
    print("üöÄ Both convolution and interpolation GPU functions linked!")
except subprocess.CalledProcessError as e:
    print(f"‚ùå Failed to compile example3_gpu_real: {e}")
    print(f"Error: {e.stderr}")
    exit(1)

print("\nüéâ COMPILATION COMPLETE!")
print("‚úÖ All components compiled successfully")
print("‚úÖ Ready to run KLT with full GPU acceleration!")
print("üöÄ GPU functions available: convolution + interpolation")
```
